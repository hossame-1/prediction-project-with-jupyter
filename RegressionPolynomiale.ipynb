```python
#importer les bibliothèques
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
#lire le fichier de données
data = pd.read_excel(r"C:\Users\HossameAhmich\Downloads\dataset_polynomial_noise_1880_2025.xlsx")
```

```python
#deviser les données en variables exolicatifs et cible
x = data.iloc[0:,:-1]
y = data.iloc[0:,-1]
```

```python
#visualizer les données
plt.scatter(x, y, color = 'blue')
```

```python
#Créer le model et le entrainer
regressor = LinearRegression()
regressor.fit(x,y)
y_predict = regressor.predict(x)
```

```python
plt.scatter(x, y, color = 'blue')
plt.plot(x, regressor.predict(x), color = 'red')
```

```python
#calculer la performance du mon model
r2 = r2_score(y, y_predict)
r2
```

```python
#Créer et entrainer le modele polynomiale
poly_regress = PolynomialFeatures(degree = 3)
x_poly = poly_regress.fit_transform(x)
regressor_poly = LinearRegression()
regressor_poly.fit(x_poly, y)
y_poly_pred = regressor_poly.predict(x_poly)
```

```python
#visualizer les données
plt.scatter(x, y, color = 'blue')
plt.plot(x, regressor.predict(x), color = 'red')
plt.plot(x, regress_poly.predict(x_poly), color = 'green')
```

```python
#calculer l performance
r2 = r2_score(y, y_poly_pred)
r2
```
